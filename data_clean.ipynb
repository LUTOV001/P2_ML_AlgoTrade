```python
#Import
import pandas as pd
from pathlib import Path
import datetime as dt
from finta import TA
```

```python
#read data to df
dia_df = pd.read_csv(Path('./Resources/DIA_data.csv'), index_col = "Date", parse_dates = True, infer_datetime_format = True)
iwm_df = pd.read_csv(Path('./Resources/IWM_data.csv'), index_col = "Date", parse_dates = True, infer_datetime_format = True)
qqq_df = pd.read_csv(Path('./Resources/QQQ_data.csv'), index_col = "Date", parse_dates = True, infer_datetime_format = True)
spy_df = pd.read_csv(Path('./Resources/SPY_data.csv'), index_col = "Date", parse_dates = True, infer_datetime_format = True)
```

```python
#SUPERFLUOUS --- #checks if dataframes were created
display(dia_df.head(2), iwm_df.head(2), qqq_df.head(2), spy_df.head(2))
```

```python
# building out returns from Close
dia_df["actual_returns"] = dia_df["Close"].pct_change()
iwm_df["actual_returns"] = iwm_df["Close"].pct_change()
qqq_df["actual_returns"] = qqq_df["Close"].pct_change()
spy_df["actual_returns"] = spy_df["Close"].pct_change()
```

```python
# dropping N/A's
dia_df.dropna(inplace=True)
iwm_df.dropna(inplace=True)
qqq_df.dropna(inplace=True)
spy_df.dropna(inplace=True)
```

```python
#SUPERFLUOUS --- #checks if dataframes were created
display(dia_df.head(2), iwm_df.head(2), qqq_df.head(2), spy_df.head(2))
```

```python
# Create a simple moving average (SMA) using a window size of 4. 
# Assign this to a column called `sma_fast`
short_window = 4
dia_df['sma_fast'] = TA.SMA(dia_df, 4)
iwm_df['sma_fast'] = TA.SMA(iwm_df, 4)
qqq_df['sma_fast'] = TA.SMA(qqq_df, 4)
spy_df['sma_fast'] = TA.SMA(spy_df, 4)
```

```python
# Create a simple moving average (SMA) using a window size of 100. 
# Assign this to a column called `sma_slow`
long_window = 100
dia_df['sma_slow'] = dia_df['Close'].rolling(window=long_window).mean()
iwm_df['sma_slow'] = iwm_df['Close'].rolling(window=long_window).mean()
qqq_df['sma_slow'] = qqq_df['Close'].rolling(window=long_window).mean()
spy_df['sma_slow'] = spy_df['Close'].rolling(window=long_window).mean()
```

```python
# Create additional technical indicators for dia_df
dia_df["ssma"] = TA.SSMA(dia_df)
dia_df["ema"] = TA.EMA(dia_df, 50)
dia_df["dema"] = TA.DEMA(dia_df)
dia_df["tema"] = TA.TEMA(dia_df)
dia_df["trima"] = TA.TRIMA(dia_df)
```

```python
# Create additional technical indicators for iwm_df
iwm_df["ssma"] = TA.SSMA(iwm_df)
iwm_df["ema"] = TA.EMA(iwm_df, 50)
iwm_df["dema"] = TA.DEMA(iwm_df)
iwm_df["tema"] = TA.TEMA(iwm_df)
iwm_df["trima"] = TA.TRIMA(iwm_df)
```

```python
# Create additional technical indicators for iwm_df
qqq_df["ssma"] = TA.SSMA(qqq_df)
qqq_df["ema"] = TA.EMA(qqq_df, 50)
qqq_df["dema"] = TA.DEMA(qqq_df)
qqq_df["tema"] = TA.TEMA(qqq_df)
qqq_df["trima"] = TA.TRIMA(qqq_df)
```

```python
# Create additional technical indicators for iwm_df
spy_df["ssma"] = TA.SSMA(spy_df)
spy_df["ema"] = TA.EMA(spy_df, 50)
spy_df["dema"] = TA.DEMA(spy_df)
spy_df["tema"] = TA.TEMA(spy_df)
spy_df["trima"] = TA.TRIMA(spy_df)
```

```python
# dropping N/A's
dia_df.dropna(inplace=True)
iwm_df.dropna(inplace=True)
qqq_df.dropna(inplace=True)
spy_df.dropna(inplace=True)
```
